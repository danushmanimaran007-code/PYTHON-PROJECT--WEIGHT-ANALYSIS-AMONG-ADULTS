```python
df['YEAR'] = df['YEAR'].str.extract(r'(\d{4})')[0].astype(int)
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='YEAR', y='ESTIMATE', hue='AGE')
plt.title('Weight Estimated Over the Years by Age Group')
plt.xlabel('Year')
plt.ylabel('Weight Estimate')
plt.legend(title='Age Group')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='AGE', y='ESTIMATE')
plt.title('Weight Estimates by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Weight Estimate')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=[np.number])
correlation_matrix = numeric_df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
